## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

### ① 印出最新文章的「作者」「標題」「時間」

```python
print('title:',soup.find_all(class_="title")[-6].text.replace('\t','').replace('\n',''))
print('author:',soup.find_all(class_="author")[-6].text.replace('\t','').replace('\n',''))
print('time:', soup.find_all(class_="date")[-6].text.replace('\t','').replace('\n',''))
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
title = []
author = []
datetime = []
for d in soup.find_all(class_="title"):
    title.append(d.text.replace('\t', '').replace('\n', ''))
    print(d.text.replace('\t', '').replace('\n', ''))
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html5lib")
        author.append(r.find(class_='article-meta-value').text)
        datetime.append(r.find_all(class_='article-meta-value')[-1].text)
        print('作者: ' + r.find(class_='article-meta-value').text, ' 時間:' + r.find_all(class_='article-meta-value')[-1].text)
    except:
        author.append(np.nan)
        datetime.append(np.nan)
        continue
```

#### **練習**
* strptime
* dictionary to dataframe

```python
url = 'https://www.ptt.cc/bbs/Gossiping/index.html'
r = requests.get(url,cookies={'over18':'1'})

soup = BeautifulSoup(r.text, "html5lib")

posts = []

for d in soup.find_all(class_="title"):
    post = {}
    post['title'] = d.text.replace('\t', '').replace('\n', '')
    try:
        r = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href'],cookies={'over18':'1'}).text, "html5lib")
        post['author'] = r.find_all(class_='article-meta-value')[0].text
        post['date'] = r.find_all(class_='article-meta-value')[3].text
    except:
        continue
    post['date'] = datetime.datetime.strptime(post['date'], "%a %b  %d %H:%M:%S %Y")
    #print(post['date'])
    posts.append(post)


posts = sorted(posts, key= lambda x: x['date']) #########

for p in posts:
    print(p['title'],p['author'],p['date'])
```

```python
df = pd.DataFrame([posts[i] for i in range(len(posts))])
```
