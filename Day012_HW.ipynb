### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd
date_list = []
type_list = []
title = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    date_list.append(d.find(class_="date").text)
    type_list.append(d.find(class_ = 'tag').text)
    title.append(d.find_all('a')[-1].text)
    
news_dict = {'date':date_list,'type':type_list,'title':title}
news = pd.DataFrame(news_dict)
```

### ③ 哪一個類別的文章最多
- 體育類別

```python
news.groupby('type')[['title']].count().sort_values(by = 'title',ascending = False)
```
