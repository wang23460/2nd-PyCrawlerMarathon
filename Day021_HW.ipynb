## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
from datetime import datetime, timedelta
```

```python
# 打開瀏覽器
browser = webdriver.Chrome(executable_path='chromedriver')
browser.maximize_window()
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
    try:
        browser.find_element_by_xpath('/html/body/div[1]/button[2]').click()
    except:
        pass
```

```python
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞
* strptime

```python
from datetime import datetime, timedelta

now = datetime.now()
past = now - timedelta(hours=2)
print("time", now)
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    news_time = datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
    if news_time > past:    
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    else:
        pass
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium.webdriver.support.ui import Select
selectSite = Select(browser.find_element_by_id("selD"))
selectSite.select_by_value('14')
browser.find_element_by_id('button').click()
```

```python
for i in range(20):
    time.sleep(3)
    browser.execute_script("window.scrollTo(0, 20000);")
    try:
        browser.find_element_by_xpath('/html/body/div[1]/button[2]').click()
    except:
        pass
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    news_hour = datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M').timetuple()[3]
    if 17 >= news_hour >= 15:    
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
    else:
        pass
```
