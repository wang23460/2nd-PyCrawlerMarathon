```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
data = table.find_all('tr')
```

```python
import re

i, j = 0, 0
dict_ = {}

air = table.find_all(name = 'tr')
# 氣體種類
for i in range(10):
    # 月份
    for j in range(12):
        # 將該氣體的當月所有資料放到data裡
        data = re.search(re.compile('([A-Z]+\d*||[A-Z]+\w)?(ppm||μg/m3||ppb)?(\d{4}/\d{2})(\d+\.?\d*)'), 
                         air[i * 12 + j + 1].text)
        
        # 將氣體名稱放到d變數裡
        if j == 0:
            gas = data.group(1)
            dict_[gas] = {}
        
        # 將當月氣體的數值放入d變數裡
        month = data.group(3)
        value = data.group(4)
        dict_[gas][month] = value
print(dict_)
```
